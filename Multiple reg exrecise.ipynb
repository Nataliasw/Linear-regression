```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
data = pd.read_csv('real_estate_multiple_reg.csv')
data
```

```python
data.describe()
```

```python
y = data['price']
x1 = data[['size','year']]
```

```python
plt.scatter(x1,y)
plt.xlabel('size',fontsize=20)
plt.ylabel('price',fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
