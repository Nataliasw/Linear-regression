```python
data_cleaned = data_cleaned.drop(['Price'], axis=1)
```

## Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])] 
vif['features'] = variables.columns
```

```python
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

## Create dummy variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
```

```python
data_with_dummies.head()
```

```python
variable = data_with_dummies[['Mileage','EngineV','log_price','Brand_BMW','Brand_Mercedes-Benz','Brand_Mitsubishi','Brand_Renault','Brand_Toyota','Brand_Volkswagen']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variable.values,1) for i in range(variable.shape[1])]
vif['features'] = variable.columns
```

```python
vif
```

## Rearrange a bit

```python
data_with_dummies.columns.values
```

```python
cols =['log_price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
```

```python
data_preprocessed.head()
```

```python
variables = data_preprocessed
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
vif
```

## Linear regression model

### Declare the inputs and the targets

```python
target = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis = 1)
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, target, test_size=0.2, random_state=365)
```

### Create the regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.show()
```

```python
sns.displot(y_train - y_hat, kde=True, binwidth=0.1)
plt.title('Residuals PDF', size=15)
```

```python
reg.score(x_train,y_train)
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

## Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_train,y_hat, alpha=0.2)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residuals'] = df_pf['Target'] - df_pf['Prediction']
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residuals']/df_pf['Target']*100)
```

```python
df_pf
```
